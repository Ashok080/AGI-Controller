```python
!pip install datasets

from datasets import load_dataset

datasets = []

# Load ~1% of HackerNews
hn_dataset = load_dataset('EleutherAI/pile', 'hacker_news', split='train[:5%]')
datasets.append(hn_dataset)

# Load ~1% of GitHub text
github_dataset = load_dataset('EleutherAI/pile', 'github', split='train[:5%]')
datasets.append(github_dataset)

# Merge and save
with open('pile_hn_github_sample.txt', 'w', encoding='utf-8') as f:
    for ds in datasets:
        for item in ds:
            f.write(item['text'] + '\n')

!ls -lh pile_hn_github_sample.txt
```
